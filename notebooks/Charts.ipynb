```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.colors import rgb2hex
```

```python
data = pd.read_csv('exp2_results_10features.csv')
```

```python
data
```

```python
x_axis = 'Unnamed: 0'
y_axes = ['G-mean']

fig, ax = plt.subplots()
for i, col in enumerate(y_axes):
    ax.bar(data[x_axis], data[col], label=col, bottom=data[y_axes[:i]].sum(axis=1))
    for j, val in enumerate(data[col]):
        ax.text(j, data[y_axes[:i]].sum(axis=1)[j] + val/2, str(round(val, 2)), ha='center', va='bottom')
ax.set_xticklabels(data[x_axis], rotation=45, ha='right')
ax.legend()
# Add a title to the chart
ax.set_title('Comparison of Model Performance in terms of Gmean')

# Save the chart as a PNG file
fig.savefig('exp2_results_10features.png', dpi=2000, bbox_inches='tight')
# Show the chart
plt.show()
```

```python
data = pd.read_csv('exp3_results_10features.csv')
data
```

```python

x_axis = data.columns[0]
y_axis = 'G-mean'

# Filter the data to select the first 8 rows
filtered_data = data.iloc[0:8]

# Create a line plot with the data
fig, ax = plt.subplots(figsize=(10, 6))
model_data = filtered_data[y_axis].values
colors = plt.cm.viridis(np.linspace(0.3, 1, len(model_data)))
for i in range(len(model_data)):
    color_hex = rgb2hex(colors[i][:3])
    ax.plot([i], [model_data[i]], 'o', markersize=8, label=filtered_data.iloc[i][x_axis], color=color_hex)
    ax.text(i+0.2, model_data[i]+0.005, f"{model_data[i]:.4f}", ha='center', va='bottom', fontsize=9)

ax.plot(model_data, linestyle='--', linewidth=2, color='darkorange', label='Average G-mean')
ax.set_xlabel('Sampling ratio', fontsize=12)
ax.set_ylabel(y_axis, fontsize=12)
ax.set_title('Performance of Balanced Random Forest with sampling', fontsize=14)
ax.set_xticks(range(len(model_data)))
ax.set_xticklabels(['1000:1', '100:1', '10:1', '4:1', '3:1', '2:1', '4:3', '1:1'])
ax.legend()
fig.tight_layout()

# Save the chart as a PNG
fig.savefig('exp3_results_10features_BRF.png', dpi=2000, bbox_inches='tight')
# Show the chart
plt.show()
```

```python

x_axis = data.columns[0]
y_axis = 'G-mean'

filtered_data_WRF = data.iloc[16:24]
filtered_data_RF = data.iloc[8:16]
filtered_data_BRF = data.iloc[0:8]

# Create a single line plot with the data
fig, ax = plt.subplots(figsize=(10, 6))

def plot_data(model_data, filtered_data, line_label, color_range, line):
    colors = plt.cm.viridis(np.linspace(color_range[0], color_range[1], len(model_data)))
    for i in range(len(model_data)):
        color_hex = rgb2hex(colors[i][:3])
        ax.plot([i], [model_data[i]], 'o', markersize=8, color=color_hex)
        # ax.text(i+0.2, model_data[i]+0.002, f"{model_data[i]:.4f}", ha='center', va='bottom', fontsize=9)
    ax.plot(model_data, linestyle='--', linewidth=2, color=line, label=line_label)

# Plot the data for the Weighted Logistic Regression
model_data_WRF = filtered_data_WRF[y_axis].values
plot_data(model_data_WRF, filtered_data_WRF, 'Average G-mean - WRF', (0.9, 0.5), 'red')

# Plot the data for the Random Forest
model_data_RF = filtered_data_RF[y_axis].values
plot_data(model_data_RF, filtered_data_RF, 'Average G-mean - RF', (0.3, 1), 'darkcyan')

# Plot the data for the Balanced Random Forest
model_data_BRF = filtered_data_BRF[y_axis].values
plot_data(model_data_BRF, filtered_data_BRF, 'Average G-mean - BRF', (0.3, 1), 'darkorange')

ax.set_xlabel('Sampling ratio', fontsize=12)
ax.set_ylabel(y_axis, fontsize=12)
ax.set_title('Performance of Weighted Logistic Regression, Random Forest, and Balanced Random Forest with sampling', fontsize=14)
ax.set_xticks(range(len(model_data_WRF)))
ax.set_xticklabels(['1000:1', '100:1', '10:1', '4:1', '3:1', '2:1', '4:3', '1:1'])
ax.legend()
fig.tight_layout()

# Save the chart as a PNG 
fig.savefig('exp3_results_10features_all.png', dpi=2000, bbox_inches='tight')
# Show the chart
plt.show()
```
